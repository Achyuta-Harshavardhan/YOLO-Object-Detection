## Changes to be done in the config file for training:

1. Uncomment the line 5,6 and 7
2. Change the number of filters for convolution layer "[convolution]" just before every yolo output "[yolo]" such that the number of filters = 3x(5+no of classes). ===> (127,171)
3. Change the number of classes ===> (135,177)
4. Create a file object.names defining all the classes
5. Some Changes need to be done in trainer.data
6. Pretrained model can be downloaded from this link (https://pjreddie.com/media/files/darknet53.conv.74) trained on COCO dataset which is used for transfer learning.

```
!./darknet detector train trainer.data cfg/yolov3-tiny.cfg darknet53.conv.74
```
